```python
# Examples for data taken in 2022 are shown
# First, make pass list of all the observation files (refer this pass list later when we download the data)
import os, glob, re
from pathlib import Path
from tqdm import tqdm
# parent directory
root_dir = "/home/idies/workspace/headata/FTP/swift/data/obs"

# months
months = ["2022_01", "2022_02", "2022_03", "2022_04", "2022_05", "2022_06", "2022_07", "2022_08", "2022_09", "2022_10", "2022_11", "2022_12"]

# taboo target id
ban_targetids = {str(x).strip().zfill(5)[-5:] for x in seqs}

# unify *xpcw3*cl.evt.gz 
all_xpcw3 = []
for m in tqdm(months):
    month_dir = os.path.join(root_dir, m)
    all_xpcw3 += glob.glob(os.path.join(month_dir, "**", "*xpcw3po_cl.evt.gz"), recursive=True)

all_xpcw3 = sorted(set(all_xpcw3))

# extract obsid of each data to see if it is taboo
obsid_dir_re = re.compile(r"/(\d{11})/")        # .../<11桁>/...
obsid_fn_re  = re.compile(r"sw(\d{11})", re.I)  # sw<11桁>...

def banned_by_tid(path: str) -> bool:

    for m in obsid_dir_re.finditer(path):
        if m.group(1)[3:8] in ban_targetids:
            return True
   
    base = os.path.basename(path)
    for m in obsid_fn_re.finditer(base):
        if m.group(1)[3:8] in ban_targetids:
            return True
    return False

kept = [p for p in all_xpcw3 if not banned_by_tid(p)]

print(f"found={len(all_xpcw3)}  kept(after exclude)={len(kept)}")
print("\n".join(kept[:5]))  # サンプル

# save paths of control data candidates in 2022
Path("xpcw3_2022_full.txt").write_text("\n".join(kept) + ("\n" if kept else ""))

# components of .txt
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010000043/xrt/event/sw00010000043xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010000044/xrt/event/sw00010000044xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010000045/xrt/event/sw00010000045xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010197015/xrt/event/sw00010197015xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010374198/xrt/event/sw00010374198xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010374199/xrt/event/sw00010374199xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010374200/xrt/event/sw00010374200xpcw3po_cl.evt.gz
# /home/idies/workspace/headata/FTP/swift/data/obs/2022_01/00010375071/xrt/event/sw00010375071xpcw3po_cl.evt.gz...
```

```python
# data selection
# only b>5° and steady 
from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.io import fits
from pathlib import Path
import numpy as np
import re
from tqdm import tqdm
SNAME_RE = re.compile(r'(?i)sw\d{8}')

def is_on_galactic_plane(ra_deg, dec_deg, b_thresh_deg=5.0):
    c = SkyCoord(ra_deg*u.deg, dec_deg*u.deg, frame='icrs')
    b = c.galactic.b.deg
    return (abs(b) <= b_thresh_deg), b

paths = [line for line in Path("/home/idies/workspace/Storage/Yu_M/persistent/xpcw3_2022_full.txt")
         .read_text().splitlines() if line.strip()]

first_obs_safe = []
m_list = []
for s in paths:
    p = Path(s)
    name = p.name
    m = re.search(r'sw(\d{8})', name)
    m=m.group(1)
    if m in m_list:
        continue
    m_list.append(m)
    
    i = next(k for k, ch in enumerate(m) if ch != '0')
    j_end_exclusive = len(m)
    key_str = m[i:j_end_exclusive]
    key_str = float(key_str)
    if (40000<=key_str<=79999) or (82000<=key_str<=87999):
        first_obs_safe.append(s)

print(len(paths))
print(len(first_obs_safe))

off_paths = []  

for s in tqdm(first_obs_safe):
    p = Path(s)

    # get RA/DEC
    with fits.open(p, memmap=False) as hdul:
        h1 = hdul["EVENTS"].header if "EVENTS" in hdul else {}
        h0 = hdul[0].header
        ra = h1.get("RA_PNT")  or h0.get("RA_PNT")
        dec= h1.get("DEC_PNT") or h0.get("DEC_PNT")

    if ra is None or dec is None:

        continue

    on_plane, bdeg = is_on_galactic_plane(float(ra), float(dec), b_thresh_deg=5.0)

    # out of galactic plane
    if not on_plane:
        off_paths.append(str(p))

# # テキストに保存（1行=1パス、末尾改行あり）
# out = Path("xpcw3_offGplane.txt")
# out.write_text("\n".join(off_paths) + ("\n" if off_paths else ""))
# print(f"total={len(paths)}  off-plane={len(off_paths)}  -> {out.resolve()}")

import shutil, gzip
kept=off_paths
# output
dest_dir = Path("/home/idies/workspace/Storage/Yu_M/persistent/eventfiles_2022")
dest_dir.mkdir(parents=True, exist_ok=True)

# copy the files (.evt.gz）
for src in kept:
    src_p = Path(src)
    dst = dest_dir / src_p.name
    shutil.copy2(src_p, dst)  # follow_symlinks=True がデフォルト

# unzip
for gz in dest_dir.glob("*.evt.gz"):
    evt = gz.with_suffix("")  # ".evt.gz" → ".evt"
    with gzip.open(gz, "rb") as fi, open(evt, "wb") as fo:
        shutil.copyfileobj(fi, fo)
    gz.unlink()  

```
